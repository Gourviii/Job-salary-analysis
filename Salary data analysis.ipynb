```python
#IQR (interquartile Range) Method

Q1 = df['Salary'].quantile(0.25)
Q3= df['Salary'].quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
```

```python
df = df[(df['Salary']>= lower_limit) & (df['Salary'] <= upper_limit)]
```

```python
df
```

# univariate analysis

```python
#salary distribution

plt.figure(figsize=(8,5))
sns.histplot(df['Salary'], bins = 30, kde=True)

plt.ticklabel_format(style='plain', axis='x')

plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Count')
plt.show()
```

# Which job roles have the highest salary?

```python
Job_salary = df.groupby('Job Roles')['Salary'].mean().sort_values(ascending=False)
```

```python
Job_salary 
```

```python
plt.figure(figsize=(8,5))


Job_salary.plot(kind='bar')
plt.title('Distributon of salary')
plt.ylabel('salary')
plt.xlabel('Job Roles')
plt.show()
    
```

# Which cities offer highest salary?

```python
Location_salary = df.groupby('Location')['Salary'].mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=(8,5))


colors = plt.cm.plasma(np.linspace(0, 1, len(Location_salary)))

Location_salary.plot(kind='bar', color=colors, edgecolor='black')
plt.title('Tp 10 cities by average salary')
plt.ylabel('Average salary')
plt.xlabel('Locations')
plt.xticks(rotation=45,ha='right')
plt.grid(axis='y' , linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

# How does average salary change as company rating increased ?
